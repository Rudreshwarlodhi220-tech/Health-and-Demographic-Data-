```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1st Load our CSV file
df = pd.read_csv('heart_disease_health_indicators_BRFSS2015 2 (1).csv')
df
```

```python
# Bar chart for Smoking Status
df['Smoker'].value_counts().plot(kind='bar', color='skyblue')

title_text = "Smoking Status Distribution"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Pie chart for Physical Activity
df['PhysActivity'].value_counts().plot(kind='pie', autopct='%1.1f%%')

title_text = "Physical Activity Distribution"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Grouped bar chart: Heart Disease vs Smoking
sns.countplot(x='Smoker', hue='HeartDiseaseorAttack', data=df)

title_text = "Heart Disease by Smoking Status"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Stacked bar chart: Heart Disease vs Physical Activity
pd.crosstab(df['PhysActivity'], df['HeartDiseaseorAttack']).plot(kind='bar', stacked=True)

title_text = "Heart Disease by Physical Activity"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Histogram for BMI
df['BMI'].plot(kind='hist', bins=30, color='orange')
"BMI Distribution"

title_text = "BMI Distribution"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Box plot for Mental Health Days
sns.boxplot(x=df['MentHlth'])

title_text = "Mental Health Days Distribution"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Correlation matrix
corr = df[['BMI','MentHlth','PhysHlth']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')

title_text = "Correlation Heatmap"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
sns.scatterplot(x='BMI', y='PhysHlth', hue='HeartDiseaseorAttack', data=df)

title_text ="BMI vs Physical Health Days"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Example: ordered categories (Age)
df.groupby('Age')['HeartDiseaseorAttack'].mean().plot(kind='line', marker='o')

title_text = "Heart Disease Trend Across Age Groups"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Violin plot
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=df)

title_text = "BMI Distribution by Heart Disease"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```

```python
# Pair plot
sns.pairplot(df[['BMI','MentHlth','PhysHlth','HeartDiseaseorAttack']], hue='HeartDiseaseorAttack')
df['HeartDiseaseorAttack'] = df['HeartDiseaseorAttack'].replace({1: 'Yes', 0: 'No'})

title_text = "xy"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
plt.show()
```
