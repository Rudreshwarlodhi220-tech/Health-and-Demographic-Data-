```python
import numpy  as ny
import pandas as pd
```

```python
df = pd.read_csv('heart_disease_health_indicators_BRFSS2015 2 (1).csv')
df
```

```python
df.columns
```

```python
df.columns == 'object'
```

```python
df.dtypes
```

```python
cols = ['HeartDiseaseorAttack', 'HighBP', 'HighChol', 'CholCheck',
        'Smoker', 'Stroke', 'PhysActivity', 'Fruits', 'Veggies',
        'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk']

df[cols] = df[cols].replace({1: 'Yes', 0: 'No'})
```

```python
df
```

```python
df['Sex'] = df['Sex'].map({0:'Female',1:'Male'})
df

# df['Sex'] = df['Sex'].replace({1:'Male',0:'Feamale'})
# df

#df['Sex'] = df['Sex'].apply(lambda x : 'Male' if x == 1 else 'Female')
#df
```

```python
df['Diabetes'] = df['Diabetes'].apply(lambda x :'No diabetes' if x == 0 
                                      else('Prediabetes' if x == 1 else 'Yes_Diabetes'))
df
```

```python
df['Education'] = df['Education'].replace({1 : 'No_sch_or_KG', 
                                           2 :'1-8 (Elementary)',
                                           3 :'9-11 (Some high school)',
                                           4 :'High_sch/Graduate)',
                                           5 :'SomeCollege/Tec(1-3)',
                                           6 :'College_Gradu(4+years)'})
```

```python
df
```

```python
df['Income'] = df['Income'].replace({1:'$2,00-$10,000',
                                    2:'$10,000-$15,000',
                                    3:'$15,000-$20,000)',
                                    4:'$20,000-$25,000)'

                                    })
df
```
