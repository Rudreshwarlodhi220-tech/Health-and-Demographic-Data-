# a. Examine Distributions of Individual Variables:
Plot histograms or bar charts to visualize the distribution of key variables such as age, BMI, mental health days (Menthlth), and physical health days (PhysHlth).

# b. Investigate Prevalence of Health Conditions:
Plot bar charts or pie charts to show the prevalence of high blood pressure (HighBP), high cholesterol (HighChol), and other health conditions.

# c. Analyze Distribution of Heart Disease (Target Variable):
Visualize the distribution of heart disease cases using a bar chart or pie chart.

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('heart_disease_health_indicators_BRFSS2015 2 (1).csv')
sns.set(style="whitegrid")
data 
```

```python
# a. Univariate Analysis- Distributions of individual variables
plt.figure(figsize=(12, 8)) # 	figsize(width : hight) - The numbers are in inches. So  means the figure will be 12 inches wide and 8 inches tall.
```

```python
# Age distribution
plt.subplot(2, 2, 1)
sns.histplot(data['Age'], bins=20, kde=True, color='skyblue')

title_text = "Distribution of Age By Rudreshwar"
plt.title(title_text)

# Save file with title as filename
filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
```

```python
# BMI distribution
plt.subplot(2, 2, 2)
sns.histplot(data['BMI'], bins=20, kde=True, color='orange')


title_text = "Distribution of BMI By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
```

```python
# Mental health days
plt.subplot(2, 2, 3)
sns.histplot(data['MentHlth'], bins=20, kde=False, color='green')

title_text = "Distribution of Mental Health Days By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
```

```python
# Physical health days
plt.subplot(2, 2, 4)
sns.histplot(data['PhysHlth'], bins=20, kde=False, color='red')

plt.tight_layout() # automatically adjusts the spacing between subplots so everything fits neatly inside the figure.
#plt.show()  # if we will write here it produces 2 graph .


title_text = "Distribution of Physical Health Days By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)

plt.show()
```

```python
# b. Univariate Analysis- Prevalence of Health Conditions
plt.figure(figsize=(12, 5)) 
```

```python
# High Blood Pressure prevalence
plt.subplot(1, 2, 1)
sns.countplot(x='HighBP', data=data, palette='Blues')
# sns.countplot(x='HighBP', data=data, palette='Blues') to remove error remove palette .


title_text = "Prevalence of Health Conditions By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)
```

```python
# High Cholesterol prevalence
plt.subplot(1, 2, 2)
sns.countplot(x='HighChol', data=data, palette='Purples')

plt.tight_layout()

title_text = "Prevalence of Health Conditions By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)

plt.show()
```

```python
# Pie chart for multiple conditions
conditions = ['HighBP', 'HighChol']
condition_counts = data[conditions].sum()

plt.figure(figsize=(6, 6))
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])

title_text = "Prevalence of Health Conditions By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)

plt.show()
```

```python
# c. Univariate Analysis- Distribution of Heart Disease (Target Variable)

plt.figure(figsize=(6, 5))
sns.countplot(x='HeartDiseaseorAttack', data=data)

title_text = "Distribution of Heart Disease Cases By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)

plt.show()
```

```python
# Bivariate Analysis 

# Age vs Heart Disease
plt.figure(figsize=(8, 6))
sns.boxplot(x='HeartDiseaseorAttack', y='Age', data=data,  hue='HeartDiseaseorAttack', palette='Set3', legend=False)
#plt.title("Age Distribution by Heart Disease Status")

title_text = "Age_Distrib_by_Heart_Disease_By_Rudreshwar"
plt.title(title_text)

# Save file with title as filename
filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)

plt.show()
```

```python
# BMI vs Heart Disease
plt.figure(figsize=(8, 6))
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=data, hue='HeartDiseaseorAttack', palette='Set1',
            legend=False)
title_text = "BMI Distribution by Heart Disease Status By Rudreshwar"
plt.title(title_text)

filename = title_text.replace(" ", "_") + ".png"
plt.savefig(filename)

```
